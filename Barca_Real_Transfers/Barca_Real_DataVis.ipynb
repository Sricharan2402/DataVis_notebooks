```python

df = df[df.Season > '2006-2007']

bar_df = df[(df['Team_to'] =='FC Barcelona')][['Season', 'Market_value','Transfer_fee']].reset_index(drop = True)
rma_df = df[(df['Team_to'] =='Real Madrid')][['Season', 'Market_value','Transfer_fee']].reset_index(drop = True)

temp = bar_df.set_index('Season')
bar_grp = list(temp['Transfer_fee'].sum(level = 0))
temp = rma_df.set_index('Season')
rma_grp = list(temp['Transfer_fee'].sum(level = 0))
```

```python
def find_per(List, cond):
    win = 0
    tot = 0
    for i in List:
        if cond == 0 and i == 'H':
            win +=1
        elif cond == 1 and i == 'A':
            win +=1
        tot +=1
    return win/tot


def calc_stat(df):
    home_temp = df.loc[(df['HomeTeam'] == 'Barcelona') | (df['HomeTeam'] == 'Real Madrid'), ['HomeTeam','FTHG','FTAG','FTR']].set_index('HomeTeam')
    away_temp = df.loc[(df['AwayTeam'] == 'Barcelona') | (df['AwayTeam'] == 'Real Madrid'), ['AwayTeam','FTHG','FTAG','FTR']].set_index('AwayTeam')
    home_goal_sum = home_temp['FTHG'].sum(level=0)
    home_goal_conc = home_temp['FTAG'].sum(level=0)
    away_goal_sum = away_temp['FTAG'].sum(level=0)
    away_goal_conc = away_temp['FTHG'].sum(level=0)
    barHGs = home_goal_sum.loc['Barcelona']
    rmaHGs = home_goal_sum.loc['Real Madrid']
    barHGc = home_goal_conc.loc['Barcelona']
    rmaHGc = home_goal_conc.loc['Real Madrid']
    barAGs = away_goal_sum.loc['Barcelona']
    rmaAGs = away_goal_sum.loc['Real Madrid']
    barAGc = away_goal_conc.loc['Barcelona']
    rmaAGc = away_goal_conc.loc['Real Madrid']
    barH = find_per(list(home_temp.loc['Barcelona']['FTR']),0)
    rmaH = find_per(list(home_temp.loc['Real Madrid']['FTR']),0)
    barA = find_per(list(away_temp.loc['Barcelona']['FTR']),1)
    rmaA = find_per(list(away_temp.loc['Real Madrid']['FTR']),1)
    return [barHGs, barAGs, (barHGs + barAGs)],[barHGc, barAGc, (barHGc + barAGc)], [rmaHGs, rmaAGs, (rmaHGs + rmaAGs)],[rmaHGc, rmaAGc, (rmaHGc + rmaAGc)], [barH, barA, (barH + barA)/2 ], [rmaH, rmaA, (rmaH + rmaA)/2]
```

```python
barGs = { 'home': [], 'away': [], 'total': []}
barGc = { 'home': [], 'away': [], 'total': []}
rmaGs = { 'home': [], 'away': [], 'total': []}
rmaGc = { 'home': [], 'away': [], 'total': []}
barW = { 'home': [], 'away': [], 'total': []}
rmaW = { 'home': [], 'away': [], 'total': []}

for i in range(7,19):
    file = "{0}_{1}.csv".format(i,i+1)
    df = pd.read_csv(file, usecols = [2,3,4,5,6])
    bgs, bgc, rgs, rgc, bw, rw = calc_stat(df)
    barGs['home'].append(bgs[0])
    barGs['away'].append(bgs[1])
    barGs['total'].append(bgs[2])
    barGc['home'].append(bgc[0])
    barGc['away'].append(bgc[1])
    barGc['total'].append(bgc[2])
    
    rmaGs['home'].append(rgs[0])
    rmaGs['away'].append(rgs[1])
    rmaGs['total'].append(rgs[2])
    rmaGc['home'].append(rgc[0])
    rmaGc['away'].append(rgc[1])
    rmaGc['total'].append(rgc[2])
    
    barW['home'].append(bw[0])
    barW['away'].append(bw[1])
    barW['total'].append(bw[2])
    rmaW['home'].append(rw[0])
    rmaW['away'].append(rw[1])
    rmaW['total'].append(rw[2])
```

# Plotting

```python
sns.set_style("dark")
ind = np.arange(len(barGc['home']))
width = 0.4

fig, ax = plt.subplots(2,2, sharex = True, figsize = (16,8))

ax[0,0].bar(ind, bar_grp, color = '#85bb65', alpha = 0.8, edgecolor = "none", label = 'Total Purchase Money')
ax[0,0].set_yticks(np.linspace(0, 500000000, 6))
ax[0,0].set_yticklabels(['','$100M ','$200M','$300M','$400M','$500M'])
ax[0,0].set_ylabel('Transfer Money (Million $)')
ax[0,0].legend(edgecolor = "none", loc = 'upper right')

ax[0,1].plot(barGs['home'],  label = 'Home', alpha = 0.70,color = '#DB0030')
ax[0,1].plot(barGs['away'], label = 'Away', alpha = 0.70,color = '#004D98')
ax[0,1].plot(barGs['total'], label = 'Total', alpha = 0.70,color = '#EDBB00')
ax[0,1].set_yticks(np.arange(0,141,20))
ax[0,1].set_ylabel('No.of Goals Scored')
ax[0,1].legend(edgecolor = "none", loc = 'lower right')

ax[1,0].plot(barW['home'],  label = 'Home', alpha = 0.70,color = '#DB0030')
ax[1,0].plot(barW['away'], label = 'Away', alpha = 0.70,color = '#004D98')
ax[1,0].plot(barW['total'], label = 'Total', alpha = 0.70,color = '#EDBB00')
ax[1,0].set_yticks(np.arange(0,1.1,.2))
ax[1,0].set_ylabel('Win Ratio', labelpad = 25.0)
ax[1,0].legend(edgecolor = "none", loc = 'lower right')

ax[1,1].plot(barGc['home'],  label = 'Home', alpha = 0.70,color = '#DB0030')
ax[1,1].plot(barGc['away'], label = 'Away', alpha = 0.70,color = '#004D98')
ax[1,1].plot(barGc['total'], label = 'Total', alpha = 0.70, color = '#EDBB00')
ax[1,1].set_yticks(np.arange(0,60,10))
ax[1,1].set_ylabel('No.of Goals Conceded')
ax[1,1].legend(edgecolor = "none", loc = 'lower right')

[[ax[i,j].xaxis.grid(True) for i in [0,1]] for j in [0,1]]
plt.xticks(ind,['07-08','08-09','09-10','10-11','11-12','12-13','13-14','14-15','15-16','16-17','17-18','18-19',])
plt.tight_layout()
plt.suptitle('FC Barcelona Detailed Performance Analysis (2007-2019)', y = 1.06, fontsize = 30)
fig.text(0.50,-0.03 ,'La Liga seasons', fontsize = 25, ha = 'center')
plt.show()
```
